```python
mean_value = df['scholarship_exam_marks'].mean()
std_value = df['scholarship_exam_marks'].std()
```

```python
print(mean_value, '.',std_value)
```

```python
highest = mean_value + 3*std_value
lowest = mean_value - 3+std_value
```

```python
outliers = df[(highest < df['scholarship_exam_marks']) | (lowest> df['scholarship_exam_marks'])]
outliers
```
