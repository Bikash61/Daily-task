```python
import pandas as pd
import numpy as np 

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')
impute.fit(df.iloc[ :,2:3 ].values)
```

```python
df.iloc[:2,3] = impute.transform(df.iloc[:,2:3])
```
