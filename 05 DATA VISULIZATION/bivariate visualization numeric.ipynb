```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

## Scatterplot

```python
sns.scatterplot(x='age', y='fare', data=df, hue='survived')
```

```python
df2 = sns.load_dataset('tips')
```

```python
df2.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df2)
```

```python
sns.scatterplot(x='total_bill', y = 'tip', hue='sex', data=df2)
```

```python
sns.scatterplot( x='total_bill',y='tip',hue='sex',style='smoker',data = df2)
```

## Joint Plot

```python
sns.jointplot(x='total_bill', y='tip',data=df2)
```

```python
sns.jointplot(x='total_bill',y='tip',hue='sex',data = df2)
```

```python
sns.jointplot(x='total_bill',y='tip',data = df2,kind='reg')
```

```python
numeric_df = df2.select_dtypes(include=['float64', 'int64'])
numeric_df.corr()
```
